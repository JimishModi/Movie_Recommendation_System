Importing thr libraries

```python
import pandas as pd      # for reading the csv file
import ast            # for extracting as stroed in JSON-like format-a string representation of a list of dictionaries 
import matplotlib.pyplot as plt   # for plotting the histogram

df = pd.read_csv('tmdb_5000_movies.csv')         # Load the CSV file

def extract_genre_names(genres_str):             # Function to extract genre names from the string
    try:
        genres = ast.literal_eval(genres_str)
        return [genre['name'] for genre in genres]
    except:
        return []
    

df['genre_names'] = df['genres'].apply(extract_genre_names)            # Apply the function to the 'genres' column
all_genres = df['genre_names'].explode()              # Explode the list so each genre gets its own row
genre_counts = all_genres.value_counts()             # Count the occurrences of each genre


# Plot the histogram
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Genre')
plt.tight_layout()
plt.show()
```
